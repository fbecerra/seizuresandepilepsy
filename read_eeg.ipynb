```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# You need to install eegtools.io from https://pypi.python.org/pypi/EEGtools
from eegtools.io import edfplus
```

```python
filename='datasets/chb01_03.edf'
edf_data = edfplus.load_edf(filename)
labels = [label.replace('.', '') for label in edf_data.chan_lab]
df = pd.DataFrame(edf_data.X.T, columns=labels, index=edf_data.time)
#print edf_data
#df.head()
```

```python
# EEG with seizure
fig, ax = plt.subplots(2, 1, figsize=(18,16))
ax[0].plot(df.index, df['FP1-F7'])
ax[0].axvline(2996, color='r', linestyle='--')
ax[0].axvline(3036, color='r', linestyle='--')

# Zoomed-in version of *same* EEG
# Dashed lines show when seizure starts and ends
ax[1].plot(df.index, df['FP1-F7'])
ax[1].axvline(2996, color='r', linestyle='--')
ax[1].axvline(3036, color='r', linestyle='--')
ax[1].set_xlim(2900, 3100)

plt.show()
```
